## Superimposing curves


- Set the display area to 6 inches wide and 3 inches high.
- Generate 2 curves in a single graph using the 'MONTH' column for the x-axis instead of the 'DATE' column:
>- One curve with 1948 values in 'red' color.
>- A curve with the 1949 values of color 'blue'.
- Use plt.show() to display the curves.


```python
fig = plt.figure(figsize=(6,3))
unrate['MONTH'] = unrate['DATE'].dt.month
plt.plot(unrate[0:12]['MONTH'], unrate[0:12]['UNRATE'], c = 'red')
plt.plot(unrate[12:24]['MONTH'], unrate[12:24]['UNRATE'], c = 'blue')
plt.show()
```

## Add more curves


- Set the graph display area to 10 inches wide and 6 inches high.
- Generate the following curves on the same graph:
>- 1948: red curve 'red'.
>- 1949: curve in 'blue' blue
>- 1950: curved in green
>- 1951: curve in orange 'orange'.
>- 1952: curved in 'black' black
- Use plt.show() to display the curves.


```python
fig = plt.figure(figsize=(10,6))
colors = ['red', 'blue', 'green', 'orange', 'black']
for i in range(5):
    start_index = i*12
    end_index = (i+1)*12
    subset = unrate[start_index:end_index]
    label = str(1948 + i)
    plt.plot(subset['MONTH'], subset['UNRATE'], c=colors[i], label=label)
    
plt.xlabel("Month")
plt.ylabel("Unemployment Rate")
plt.title("Monthly Unemployment Trends, 1948-1952")
plt.show()
```

## Add a caption


- Modify the previous code to include a legend specifying the year for each colour.
>- This means that the 1948 data curve using the red colour 'red' will have the legend '1948' as its title.
- Position the legend in the upper left corner 'upper left'.
- Display the graph.


```python
fig = plt.figure(figsize=(10,6))
colors = ['red', 'blue', 'green', 'orange', 'black']
for i in range(5):
    start_index = i*12
    end_index = (i+1)*12
    subset = unrate[start_index:end_index]
    label = str(1948 + i)
    plt.plot(subset['MONTH'], subset['UNRATE'], c=colors[i], label=label)
plt.legend(loc = 'upper left')
plt.xlabel("Month")
plt.ylabel("Unemployment Rate")
plt.title("Monthly Unemployment Trends, 1948-1952")
plt.show()
```
